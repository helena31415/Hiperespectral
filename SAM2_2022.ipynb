Para o SWIR (Infravermelho de Ondas Curtas): Intervalo de amostragem é de 10 nm.
Comparação com a Resolução Espectral e Largura Espectral a Meio Máximo (FWHM):

Resolução Espectral: Precisão de calibração espectral.
FWHM: Largura do recurso espectral, que dá uma ideia da resolução espectral efetiva.
VNIR (420 - 1000 nm)
Cálculo do Intervalo Máximo Entre Bandas:

O intervalo de amostragem é de 6,5 nm.
Comparação com Resolução Espectral e FWHM:

Resolução Espectral: 0,5 nm
FWHM: 8,1 ± 1,0 nm
A diferença máxima entre bandas (6,5 nm) é menor que a FWHM (8,1 nm ± 1,0 nm), o que sugere que o espaçamento entre bandas é menor que a largura da característica espectral. Portanto, as bandas são suficientemente próximas em relação à FWHM, e o espectro pode ser considerado quase contínuo.

SWIR (900 - 2450 nm)
Cálculo do Intervalo Máximo Entre Bandas:

O intervalo de amostragem é de 10 nm.
Comparação com Resolução Espectral e FWHM:

Resolução Espectral: 1 nm
FWHM: 12,5 ± 1,5 nm
A diferença máxima entre bandas (10 nm) é comparável à FWHM (12,5 nm ± 1,5 nm). Embora o intervalo entre bandas seja relativamente grande em comparação com a FWHM, é aceitável para muitos tipos de análise espectral. Portanto, ainda é razoável considerar a continuidade espectral dentro desta faixa.


```python
# Define VNIR and SWIR wavelength ranges
def calculate_wavelengths(start, interval, count):
    return [start + interval * i for i in range(count)]

# Calculate VNIR and SWIR wavelengths
vnir_wavelengths = calculate_wavelengths(420, 6.5, 90)  # 90 bands for VNIR
swir_wavelengths = calculate_wavelengths(900, 10, 156)

# Combine VNIR and SWIR wavelengths
all_wavelengths = vnir_wavelengths + swir_wavelengths

# Create a mapping from band names to wavelengths
band_names = [f'b{i+1}' for i in range(len(all_wavelengths))]
wavelength_mapping = dict(zip(band_names, all_wavelengths))

# Rename the rows in the transposed DataFrame
poligonos_transposed.index = poligonos_transposed.index.map(wavelength_mapping)

# Print the first few rows of the DataFrame to verify the changes
poligonos_transposed
```

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine

# Load data (Replace with actual loading method if needed)
# filtered_owts_aggregated = pd.read_csv('path_to_filtered_owts_aggregated.csv', index_col=0)
# filtered_owts = pd.read_csv('path_to_filtered_owts.csv')

# Prepare result DataFrame
# Initialize with the same rows as filtered_owts and columns as filtered_owts_aggregated
results = pd.DataFrame(index=filtered_owts['Bandas'], columns=filtered_owts_aggregated.index)

# Align columns of both DataFrames
owts_columns = filtered_owts_aggregated.columns
poligonos_columns = filtered_owts.columns[1:]  # Exclude 'Bandas'

# Ensure columns are in the same order
assert set(owts_columns) == set(poligonos_columns), "Column names do not match!"

# Calculate SAM (Cosine Distance)
for i, poly_row in filtered_owts.iterrows():
    poly_spectrum = poly_row[1:].values  # Exclude 'Bandas' column
    for j, owt_row in filtered_owts_aggregated.iterrows():
        owt_spectrum = owt_row.values
        # Compute cosine distance
        distance = cosine(poly_spectrum, owt_spectrum)
        results.loc[poly_row['Bandas'], owt_row.name] = distance

# Determine the class
results['Class'] = results.idxmin(axis=1)
results = results.reset_index().rename(columns={'index': 'Bandas'})

# Save results to CSV
results.to_csv('SAM_results.csv', index=False)

# Display the results
print(results.head())
```
